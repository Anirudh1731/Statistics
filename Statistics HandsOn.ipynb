```python
df2=pd.read_csv('cars')
```

```python
df2
```

```python
#range
print(df2["mpg"].max()-df2["mpg"].min())
```

```python
#percentile
#quartile
#IQR
# (90 percentile of your entire dta)

# so outof entire data 90% of value has value less than 30.090
df2["mpg"].quantile(0.90)
```

```python
#25 percentile (q1)
Q1=df2["mpg"].quantile(0.25)
Q1
```

```python
# 50 percentile (q2)
# (50 % value are greater than 19.2 while 50 % are less than 19.2)
#hence this acts like median
Q2=df2["mpg"].quantile(0.50)
```

```python
#75 percentile (q3)
Q3=df2["mpg"].quantile(0.75)
Q3
```

```python
#lets verify if q2 represnts median or not

len(df2[df2["mpg"]<19.2])
```

```python
len(df2[df2["mpg"]>19.2])
```

```python
#hence 19.2 is the middle value or median
```

```python
#inter quartile range

IQR=Q3-Q1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(df2["mpg"])
```

```python
upper_limit= Q3+ 1.5*IQR
```

```python
lower_limit= Q1-1.5*IQR
```

```python
(upper_limit,lower_limit)
```

```python
# so any value aove upper_limit and any value below lower value are outliers 
```
